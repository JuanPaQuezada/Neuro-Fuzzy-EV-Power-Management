# Membership function visualization
# Neuro-Fuzzy EV power management system

### Imports

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

### Universe of discourse

```python
#State of charge (normalized)
soc=np.linspace(0,100,1000)
#Battery temperature °C
temp=np.linspace(20,120,1000)
#Driving style index
drive=np.linspace(0,1,1000)
#SOP output
sop=np.linspace(0,1,1000)
```

### Soc membership functions

```python
soc_low=fuzz.trapmf(soc,[0.0,0.0,0.2,0.4])
soc_normal=fuzz.trimf(soc,[0.25,0.55,0.85])
soc_high=fuzz.trapmf(soc,[0.7,0.85,1.0,1.0])
```

#### Plot Soc

```python
plt.figure()
plt.plot(soc, soc_low, label='Low')
plt.plot(soc, soc_normal, label='Normal')
plt.plot(soc,soc_high,label='High')
plt.title('SoC membership functions')
plt.xlabel('State of charge')
plt.ylabel('membership degree')
plt.legend()
plt.grid(True)
plt.show()
```

### Temperature membership functions

```python
temp_cold=fuzz.trapmf(temp, [20,20,25,35])
temp_ideal=fuzz.trimf(temp, [30,40,50])
temp_hot=fuzz.trapmf(temp, [45,60,120,120])
```

#### Plot temperature

```python
plt.figure()
plt.plot(temp, temp_cold, label='Cold')
plt.plot(temp, temp_ideal, label='Ideal')
plt.plot(temp, temp_hot, label='Hot')
plt.title('Battery Temperature Membership Functions')
plt.xlabel('Temperature (°C)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()
```

### Driving style membership functions

```python
drive_smooth = fuzz.trapmf(drive, [0.0, 0.0, 0.2, 0.4])
drive_normal = fuzz.trimf(drive, [0.3, 0.5, 0.7])
drive_aggressive = fuzz.trapmf(drive, [0.6, 0.8, 1.0, 1.0])
```

#### Plot driving style

```python
plt.figure()
plt.plot(drive, drive_smooth, label='Smooth')
plt.plot(drive, drive_normal, label='Normal')
plt.plot(drive, drive_aggressive, label='Aggressive')
plt.title('Driving Style Membership Functions')
plt.xlabel('Driving Style Index')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
plt.show()
```
